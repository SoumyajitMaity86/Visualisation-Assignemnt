```python
#4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def generate_and_plot_bar_chart():
    # Generate a dataset
    categories = ['A', 'B', 'C', 'D', 'E']
    values = {
        'A': np.random.normal(10, 2, 100),
        'B': np.random.normal(20, 5, 100),
        'C': np.random.normal(30, 10, 100),
        'D': np.random.normal(40, 15, 100),
        'E': np.random.normal(50, 20, 100)
    }

    # Calculate the average value for each category
    avg_values = {cat: np.mean(values[cat]) for cat in categories}

    # Convert to DataFrame for easy plotting
    df = pd.DataFrame(list(avg_values.items()), columns=['Category', 'Average Value'])

    # Plot a bar chart
    plt.figure(figsize=(10, 6))
    plt.bar(df['Category'], df['Average Value'], color='skyblue', edgecolor='black')
    plt.xlabel('Category')
    plt.ylabel('Average Value')
    plt.title('Average Value Across Categories')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# Run the function to create and display the bar chart
generate_and_plot_bar_chart()

                                                #Using a Box Plot

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def generate_and_plot_box_plot():
    # Generate a dataset
    categories = ['A', 'B', 'C', 'D', 'E']
    values = {
        'A': np.random.normal(10, 2, 100),
        'B': np.random.normal(20, 5, 100),
        'C': np.random.normal(30, 10, 100),
        'D': np.random.normal(40, 15, 100),
        'E': np.random.normal(50, 20, 100)
    }

    # Convert to DataFrame for easy plotting
    data = []
    for cat in categories:
        data.extend([(cat, val) for val in values[cat]])
    df = pd.DataFrame(data, columns=['Category', 'Value'])

    # Plot a box plot
    plt.figure(figsize=(10, 6))
    df.boxplot(column='Value', by='Category', grid=False, patch_artist=True)
    plt.suptitle('')  # Suppress the default title to show custom title
    plt.title('Distribution of Values Across Categories')
    plt.xlabel('Category')
    plt.ylabel('Value')
    plt.show()

# Run the function to create and display the box plot
generate_and_plot_box_plot()

```

```python
#5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def generate_and_plot_correlation_heatmap():
    # Generate synthetic dataset with correlated features
    np.random.seed(0)  # For reproducibility
    num_samples = 100
    feature_1 = np.random.normal(0, 1, num_samples)
    feature_2 = 0.5 * feature_1 + np.random.normal(0, 1, num_samples)
    feature_3 = 0.8 * feature_1 + 0.2 * feature_2 + np.random.normal(0, 1, num_samples)
    
    # Create a DataFrame
    df = pd.DataFrame({
        'Feature 1': feature_1,
        'Feature 2': feature_2,
        'Feature 3': feature_3
    })
    
    # Compute the correlation matrix
    corr_matrix = df.corr()
    
    # Plot the heatmap
    plt.figure(figsize=(8, 6))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f', linewidths=0.5)
    plt.title('Correlation Matrix Heatmap')
    plt.show()

# Run the function to create and display the heatmap
generate_and_plot_correlation_heatmap()
```

```python
#5. Create a Bokeh heatmap using the provided dataset.
'''data_heatmap np.random.rand(10, 10)
X np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy пр.meshgrid(x, y)'''

import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_file
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256
from bokeh.models import ColorBar
from bokeh.layouts import column

# Generate random data for heatmap
np.random.seed(30)  # For reproducibility
data_heatmap = np.random.rand(10, 10)

# Create meshgrid for X and Y axes
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten the arrays for plotting
x_flat = xx.flatten()
y_flat = yy.flatten()
z_flat = data_heatmap.flatten()

# Convert to DataFrame
df = pd.DataFrame({'x': x_flat, 'y': y_flat, 'z': z_flat})

# Set up the output file (to save the plot)
output_file("heatmap.html")

# Create a new plot with title and axis labels
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y', 
           x_axis_type='linear', y_axis_type='linear', 
           tools="", tooltips=[("X", "@x"), ("Y", "@y"), ("Z", "@z")])

# Create a color mapper
mapper = linear_cmap(field_name='z', palette=Viridis256, low=min(z_flat), high=max(z_flat))

# Add a rectangle renderer for the heatmap
p.rect(x='x', y='y', width=1/10, height=1/10, source=df, 
       line_color=None, fill_color=mapper)

# Add color bar
color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0,0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
